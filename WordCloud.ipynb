```
import wikipedia
```

```
def GetWikipediaArticle(query):
  title = wikipedia.search(query)[0]
  page = wikipedia.page(title)
  return page.content
```

```
print(GetWikipediaArticle("Batman"))
```

```
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```
def CreateWordCloud(text):
  maskImage = np.array(Image.open('batman-logo.png'))
  stopwordSet = set(STOPWORDS)
  wc = WordCloud(background_color="grey", mask=maskImage, max_words=100, stopwords=stopwordSet)
  wordCloud = wc.generate(text)
  
  plt.imshow(wordCloud, interpolation='bilinear')
  plt.axis("off")
  
  wc.to_file(os.path.join('/content/sample_data', 'wc.png'))
```

```
CreateWordCloud(GetWikipediaArticle("Batman"))
```
